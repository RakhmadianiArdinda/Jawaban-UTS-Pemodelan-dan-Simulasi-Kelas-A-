```python
X = data.drop(columns=['Ice Cream Sales (units)'],axis=1)
```

```python
X.head()
```

```python
y = data['Ice Cream Sales (units)']
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression 
linreg = LinearRegression()
linreg.fit(X,y)
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X, y, color = 'green')
plt.plot(X, linreg.predict(X), color = 'blue')
plt.title("Linear Regression")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
model_poly = PolynomialFeatures(degree=2)
X_poly = model_poly.fit_transform(X)
model_poly.fit(X_poly, y)
linreg2 = LinearRegression()
linreg2.fit(X_poly, y)
```

```python
plt.scatter(X, y, color = 'green')
plt.plot(X,linreg2.predict(X_poly), color = 'blue')
plt.title("Polynomial Regression dengan Degree = 2")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.show()
```
