```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint # menyelesaikan persamaan diferensial secara numerik

# Model logistik
def logistic_model(P, t, r, K):
    return r * P * (1 - P / K)

# Parameter numerik
r = 0.08      # laju pertumbuhan = 8% per tahun
K = 5000      # daya dukung = 5000 orang
P0 = 200      # populasi awal = 200 orang
t = np.linspace(0, 100)  # waktu dalam tahun dari tahun ke-0 sampai ke-100

# Solusi
sol = odeint(logistic_model, P0, t, args=(r, K)) # output dari "sol" berisi nilai populasi P(t) untuk setiap waktu t

# Visualisasi
plt.figure(figsize=(8,5))
plt.plot(t, sol, label="Populasi", color="blue")
plt.axhline(K, linestyle="--", color="red", label="Carrying Capacity (5000 orang)")
plt.title("Pertumbuhan Populasi Logistik (r = 0.08 per tahun, K = 5000 orang)")
plt.xlabel("Waktu (tahun)")
plt.ylabel("Jumlah Penduduk (orang)")
plt.legend()
plt.grid(True)
plt.show()
```
